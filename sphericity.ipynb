```python
import trimesh
import os
import pandas as pd
import numpy as np
```

```python
folder_path = "D:\\jupyter notebook\\3D model\\正式数据\\OBJ"
```

```python
model_files = [f for f in os.listdir(folder_path) if f.endswith('.obj')]
```

```python
results_df = pd.DataFrame(columns=["Model", "Sphericity"])
```

```python
for model_file in model_files:
    # 读取模型文件
    model_path = os.path.join(folder_path, model_file)
    model = trimesh.load(model_path)
    
    # 计算球形度
    volume = model.volume
    surface_area = model.area
    sphericity = (36 * np.pi * volume ** 2) ** (1 / 3) / surface_area
    
    # 将结果添加到DataFrame中
    results_df = results_df.append({"Model": model_file, "Sphericity": sphericity}, ignore_index=True)
```

```python
results_df.to_csv("Sphericity.csv", index=False)
```
